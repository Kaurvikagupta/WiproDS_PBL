```python
numbers = [10, -5, 8, -12, 20, -7, 15, -9, 30, -1]

try:
    index = int(input("Enter index (0-9): "))

    value = numbers[index]

    if value >= 0:
        print(value, "is Positive")
    else:
        print(value, "is Negative")

except IndexError:
    print("Error: Invalid index.")

except ValueError:
    print("Error: Please enter a valid integer.")
```
