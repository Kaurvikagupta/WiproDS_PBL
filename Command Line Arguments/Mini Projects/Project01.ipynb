```python
# Calculate happiness using command line arguments (Jupyter version)

import sys

# Simulate command line arguments
sys.argv = ["Project01.py", "3-1", "5-7", "1-5-3-8"]

like = sys.argv[1].split("-")
dislike = sys.argv[2].split("-")
numbers = sys.argv[3].split("-")

happiness = 0

for num in numbers:
    if num in like:
        happiness += 1
    elif num in dislike:
        happiness -= 1

print("Final Happiness:", happiness)
```
