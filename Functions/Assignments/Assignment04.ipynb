```python
# Write a function that accepts a string and prints the number of upper case and lower case letters.

def count_letters(string):
    upper = 0
    lower = 0

    for ch in string:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Upper Case Letters:", upper)
    print("Lower Case Letters:", lower)

text = input("Enter a string: ")

count_letters(text)
```
